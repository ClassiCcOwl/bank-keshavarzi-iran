```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd

from time import sleep
import warnings


warnings.filterwarnings('ignore')

option = webdriver.ChromeOptions()
option.add_argument('headless')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=option)

# driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
url = "https://www.bki.ir/%D8%B4%D8%B9%D8%A8/%D8%AC%D8%B3%D8%AA%D8%AC%D9%88-%D8%B4%D8%B9%D8%A8#"
driver.get(url)
search_btn_xpath = '//*[@id="dnn_ctr1602_View_btnSearch"]'
WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, search_btn_xpath))).click()
```

```python
view_btn_xpath = '//table//a[@title]'
df_all = pd.DataFrame()

for page in range(2,177):    
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, view_btn_xpath)))
    branches = driver.find_elements(By.XPATH, view_btn_xpath)
    for n, branch in enumerate(branches):
        sleep(1)
        branch.click()
        info_xpath = '//div[contains(@class, "jconfirm-box-")]//div[@class="form-horizontal"]//div[@class="form-group"]//div'
        WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, info_xpath)))
        information = driver.find_elements(By.XPATH, info_xpath)
        m = dict()
        for i , v in enumerate(information):
            if i%2 == 0:
                latest_key = v.text.replace(":","")
            else:
                lates_value = v.text
                m.update({latest_key:lates_value})
        with open(f'backup.txt', 'a', encoding="utf-8") as file:
            for k, v in m.items():
                file.write(str(k) +" : " + str(v) +'\n')

        df = pd.DataFrame(m, index=[0])
        df_all = df_all.append(df)

        cancel_btn_xpath = '//button[contains(text(),"")]'
        WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, cancel_btn_xpath))).click()

        print(f"branch {n+1} from page {page-1} added")

    print("+=============================+")
    print(f"page number {page-1} is done")
    p_btn_link = f'//td//a[contains(@href,"Page${page}")]'
    sleep(2)
    driver.find_element(By.XPATH, p_btn_link).click()
```

```python
df_all.to_excel('file.xlsx', encoding='utf-8', index=False)
df_all
```
